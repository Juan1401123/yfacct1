```python
allcards=driver.find_elements(By.XPATH, '//div[@class="pin masonry top_off"]')
```

```python
titulo=[]
numero=[]
descripcion=[]
```

```python
for card in allcards[:len(allcards)/2]:
    try:
        card.click()
        time.sleep(2)
        page=BeautifulSoup(driver.page_source, 'html.parser')
        cardselect=page.find('div', id='ficha_main')
        cardinfo=cardselect.find('section', id='ficha_datos')
        title = cardinfo.find('h1').text.strip()
        if title:
            titulo.append(title)
        else:
            titulo.append(None)
        description = cardselect.find('section', id='ficha_texto').text.strip()
        if description:
            descripcion.append(description)
        else:
            descripcion.append(None)
        try:      
            phone = page.find('div', id='ficha_telefono').text.strip().replace('phone', '').replace('','Llamar')
        except AttributeError:
            phone = page.find('div', id='ficha_whatsapp').get('data-telefono')

        if phone:
            numero.append(phone)
        else:
            numero.append(None)

        driver.execute_script("document.elementFromPoint(200, 300).click();")
        
    except ElementClickInterceptedException:
        print(  "ElementClickInterceptedException: Retrying click")
```

```python
sample={
    'titulo': titulo,
    'numero': numero,
    'descripcion': descripcion
}
```

```python
while driver.find_element(By.ID,'loadMore'):
    driver.find_element(By.ID,'loadMore').click()
    time.sleep(2)   
```

```python
html=BeautifulSoup(driver.page_source, 'html.parser')
```

```python
allcards= html.find_all('div', class_='item')
```

```python
nombre = []
numero = []
descripcion = []
```

```python
for card in allcards:
  
    nameandnumber = card.find('img')
    alt_text = nameandnumber['alt'].split(' ')
    for index,e in enumerate(alt_text):
        if e=='Escorts':
            number= alt_text[index-1]
        
    
    name=alt_text[0]
    description=card.find('figcaption', class_='contenido-anuncio').text.replace('\n','').replace('  ','').replace('   ','').replace('    ','').replace('     ','').replace('      ','').replace('       ','').replace('\t',' ')       
    descripcion.append(description)
    nombre.append(name)
    numero.append(number)
    
  
    
    
```

```python
sample=({
    'Nombre': nombre,
    'Numero': numero,
    'Descripcion': descripcion
})
```

```python
sample=pd.DataFrame(sample)
sample.drop_duplicates(inplace=True)
sample.to_excel('panaescort_Mujer.xlsx', index=False)
sample.to_csv('panaescort_Mujer.csv', index=False)
```
