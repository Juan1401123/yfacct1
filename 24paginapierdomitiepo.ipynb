```python
descnum=0
contador=1
while descnum==0:
    driver.get('https://www.encuentra24.com/panama-es/searchresult/bienes-raices-venta-de-propiedades.'+str(contador))
    time.sleep(2)
    soup=BeautifulSoup(driver.page_source,'html.parser')
    allrecords= soup.find_all('div', class_= re.compile(r'\bd3-ad-tile.*\bd3-ads-grid__item.*\bd3-ad-tile--fullwidth.*\bd3-ad-tile--bordered.*'))
    for record in allrecords:
        try:
            preurl=record.find('a',class_='d3-ad-tile__description')['href']
            if preurl:
                url.append('https://www.encuentra24.com/'+preurl)
        except AttributeError:
            print('error')





    if len(driver.find_elements(By.XPATH, '//a[@class="d3-pagination__arrow d3-pagination__arrow--next" and @title="Siguiente"]'))>0:
        contador+=1
    else:
        descnum=1
    print(len(url))
```

```python
info=[]
numero=[]
```

```python
for index, enlace in enumerate(url[8349:]):
    try:
        driver.get(enlace)
        show_phone = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.CLASS_NAME, "show-phone"))
    )
    
        show_phone.click()
        time.sleep(2)

    except (TimeoutException, ElementClickInterceptedException) as e:
        print('no hay1')
        try:
        #   driver.get(enlace)
          show_phone = WebDriverWait(driver, 10).until(
          EC.element_to_be_clickable((By.CLASS_NAME, "see-phone"))
    )
    
          show_phone.click()
          time.sleep(2)
        except (TimeoutException, ElementClickInterceptedException) as e:
            print('no hay2')
            continue

    page=BeautifulSoup(driver.page_source,'html.parser')
    try:
        prenumero=page.find('div',class_='d3-property-info__images').text
        if prenumero:
            prenumero=page.find('span',class_='d3-button d3-property-contact__phone').text.replace('\n','')
            if prenumero=='Llamar':
                driver.quit()
                driver = webdriver.Chrome(options=options)
                time.sleep(4)
                driver.get(enlace)
                time.sleep(2)
                driver.get(enlace)
                try:
                   show_phone = WebDriverWait(driver, 10).until(
                   EC.element_to_be_clickable((By.CLASS_NAME, "show-phone")))
                   show_phone.click()
                   time.sleep(2)
                   page=BeautifulSoup(driver.page_source,'html.parser')
                   prenumero=page.find('div',class_='d3-property-info__images').text
                   if prenumero:
                     prenumero=page.find('span',class_='d3-button d3-property-contact__phone').text.replace('\n','')
                     numero.append(prenumero)

                except (TimeoutException, ElementClickInterceptedException) as e:
                    pass
                
            else:
               numero.append(prenumero)
               print('append numero')
        else:
            numero.append(None)
            print('no append numero')
    except AttributeError:
        numero.append(None)
        print('error numero')
        
    try:
        preinfo=page.find('div',class_='d3-property-info__images').text
        if preinfo:
            preinfo=preinfo.replace('\n','').replace('\xa0',' ').strip()
            preinfo = re.sub(r'\s+', ' ',preinfo).strip()
            info.append(preinfo)
            print('append info')
        else:
            info.append(None)
            print('no append info')
    except AttributeError:
        info.append(None)
        
        print('error info')
    print(index+8349)
```

```python
df=({
    'info':info,
    'numero':numero
})
```
